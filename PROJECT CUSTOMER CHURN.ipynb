# 3.Linear Regression:
**Build a simple linear model where dependent variable is ‘MonthlyCharges’ and independent variable is ‘tenure’:**
- a. Divide the dataset into train and test sets in 70:30 ratio.
- b. Build the model on train set and predict the values on test set
- c. After predicting the values, find the root mean square error
- d. Find out the error in prediction & store the result in ‘error’
- e. Find the root mean square error

```python
# Step a: Divide the dataset into train and test sets (70:30)
X =churn[['tenure']]
y =churn['MonthlyCharges']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.3,random_state=42) 
```

```python
# step b: Build the model on training data
from sklearn.linear_model import LinearRegression

lr =LinearRegression()
lr.fit(X_train,y_train)
```

```python
# Predict on the test data
y_pred =lr.predict(X_test)
```

```python
# step 3: calculate RMSE
from sklearn.metrics import mean_squared_error

rmse =np.sqrt(mean_squared_error(y_test,y_pred))
print(f'Root Mean Squared Error Is:{rmse:.2f}')
```

```python
# Step d: Calculate prediction error and store in 'error'
error =y_test - y_pred
error
```

```python
result_df =pd.DataFrame({
    'Actual':y_test,
    'Predicted':y_pred,
    'Error':error
})
result_df
```

# 4. Logistic Regression:
**Build a simple logistic regression model where dependent variable is ‘Churn’ and independent variable is ‘MonthlyCharges’:**
- a. Divide the dataset in 65:35 ratio
- b. Build the model on train set and predict the values on test set
- c. Build the confusion matrix and get the accuracy score

```python
churn.columns
```

```python
# Step a: Divide the dataset into train and test sets (65:35)
X =churn[['MonthlyCharges']] ## independent variable
y =churn['Churn'] ## dependent variable

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=.35,random_state=42)
```

```python
## step b:Build the model on training data
from sklearn.linear_model import LogisticRegression
lr =LogisticRegression()
lr.fit(X_train,y_train)
```

```python
## predict on the test data
y_pred =lr.predict(X_test)
```

```python
# step c: build the confusion matrix and accuracy score
from sklearn.metrics import confusion_matrix,accuracy_score

cm =confusion_matrix(y_test,y_pred)
print(f'Confusion Matrix:{cm}')

plt.subplots(figsize=(10,7))
sns.heatmap(cm,annot=True,cmap='Blues',fmt='d')
plt.xlabel("Predicted Value")
plt.ylabel("Actual value")
plt.title("Confusion Matrix")
plt.show()
```

# Random Forest:
**Build a Random Forest model where dependent variable is ‘Churn’ and independent variables are ‘tenure’ and ‘MonthlyCharges’:**

 a. Divide the dataset in 70:30 ratio
 
 b. Build the model on train set and predict the values on test set
 
 c. Build the confusion matrix and calculate the accuracy

```python
# Divide the dataset into train and test sets (80:20)
X =churn[['tenure','MonthlyCharges']] ## indepenedent variable
y =churn['Churn'] ## dependent variable

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=.30,random_state=42)
```

```python
# build the model
from sklearn.ensemble import RandomForestClassifier
rf =RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
## predict on test data
y_pred =rf.predict(X_test)
```

```python
# build the confusion matrix and accuracy score
from sklearn.metrics import confusion_matrix ,accuracy_score

cm =confusion_matrix(y_test,y_pred)
print(f'Confusion Matrix:{cm}')

plt.subplots(figsize=(14,7))
sns.heatmap(cm,annot=True,cmap='coolwarm',fmt='d')
plt.title("Confusion matrix")
plt.xlabel("Predicted Value")
plt.ylabel("Actual value")
plt.tight_layout()
plt.show()


accuracy =accuracy_score(y_test,y_pred)
print(f'Accuracy on RandomForestClassifier:{accuracy*100}')
```
